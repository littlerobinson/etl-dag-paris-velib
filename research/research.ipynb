```python
import logging

import requests

import pandas as pd

OPENWEATHERMAP_API = "xxx"
```

```python
logging.info("fetch_weather_data")
lat = "48.866667"
lon = "2.333333"
headers = {
    "Accept-Encoding": "application/json",
    "User-Agent": "Chrome/126.0.0.0 ",
}

url = f"https://api.openweathermap.org/data/3.0/onecall?lat={lat}&lon={lon}&units=metric&appid={OPENWEATHERMAP_API}"
response = requests.get(url, headers=headers)
weather_data = response.json()

print(weather_data['current'])
```

```python
headers = {
    "Accept-Encoding": "application/json",
    "User-Agent": "Chrome/126.0.0.0 ",
}

url = "https://velib-metropole-opendata.smovengo.cloud/opendata/Velib_Metropole/station_status.json"
response = requests.get(url, headers=headers)
status_data = response.json()
status_data

```

```python
stations = status_data["data"]["stations"]
stations_df = pd.json_normalize(stations)
stations_df["id"] = stations_df["station_id"].astype(str) + "_" + stations_df["last_reported"].astype(str)

stations_df
```
